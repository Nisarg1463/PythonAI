```python
import numpy as np
import pandas as pd
```

```python
def date_generator(date, month, year, number_of_dates):
    dates = []
    number_of_days = {1:31, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
    for _ in range(number_of_dates):
        dates.append(f'{date}/{month}/{year}')
        
        date += 1
        leap_year = False
        if month == 2:
            if year % 4 == 0 or year % 100 == 0 :
                if year % 400 != 0 :
                    leap_year = True

        if month == 2:
            if leap_year:
                if date > 29 :
                    date = 1
                    month += 1
            else:
                if date > 28 :
                    date = 1
                    month += 1
        else:
            if date > number_of_days[month]:
                date = 1
                month += 1

        if month > 12:
            month = 1
            year += 1

    return dates
    
```
